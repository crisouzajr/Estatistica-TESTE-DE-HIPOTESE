```python
import random
import numpy as np
import pandas as pd
```

# Week 1

```python
week1 = random.sample(range(1100,1890),20)
week1
```

```python
dict_1 = {"week_1": week1}

df1 = pd.DataFrame(dict_1)

df1.to_csv('week_1.csv')
```

```python
from pandas.io.parsers import read_csv
F = read_csv('week_1.csv')
F
```

```python
week_1 = F.drop('Unnamed: 0', axis=1)
week_1
week_1.to_csv('week_1.csv')
```
